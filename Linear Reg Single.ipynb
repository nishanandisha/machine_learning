```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("Price.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('Prices')
plt.scatter(df.area,df.price,color="red",marker='+')
```

```python
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
