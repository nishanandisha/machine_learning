```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('penguins')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#one Hot Encoding
df.sex.unique()
```

```python
sex=pd.get_dummies(df['sex'],drop_first=True)
sex.head()
```

```python
df.island.unique()
```

```python
island=pd.get_dummies(df['island'],drop_first=True)
island.head()
```

```python
#concate 2 df
```

```python
new_data=pd.concat([df,island,sex],axis=1)
```

```python
new_data.head()
```

```python
new_data.drop(['sex','island'],inplace= True,axis=1)
```

```python
new_data.head()
```

```python
y=new_data.species
y.head()
```

```python
y.unique()
```

```python
y=y.map({'Adelie':0,'Chinstrap':1,'Gentoo':2})
y.head()
```

```python
new_data.drop(['species'],inplace=True,axis=1)
```

```python
x=new_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier =RandomForestClassifier(n_estimators=5,criterion='entropy',random_state=0)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#using gini criteria
classifier2 =RandomForestClassifier(n_estimators=7,criterion='gini',random_state=0)
classifier2.fit(x_train,y_train)
```

```python
y_pred2=classifier2.predict(x_test)
```

```python
accuracy_score(y_pred2,y_test)
```
